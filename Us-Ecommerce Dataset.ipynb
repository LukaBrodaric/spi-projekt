##### ER DIJAGRAM

![ER DIJAGRAM](ER_dijagram.png "ER DIJAGRAM")



```python
customer_df = df[['Gender', 'Customer_Login_type', 'customer_id']].drop_duplicates().reset_index(drop=True)
customer_df
```

```python
product_df = df[['Product', 'Category', 'Individual_Price_US$']].drop_duplicates().reset_index(drop=True)
product_df['product_id'] = product_df.index 
product_df
```

```python
state_df = df[['State', 'City']].drop_duplicates().reset_index(drop=True)
state_df['state_id'] = state_df.index 
state_df
```

```python
order_df=pd.merge(df, product_df, on=['Product', 'Category', 'Individual_Price_US$'])
order_df=pd.merge(order_df, state_df, on=['State', 'City'])
order_df = order_df[['Transaction_id','Date', 'Delivery_Type', ' Quantity ', 'Device_Type', 'Amount US$', 'Time', 'Transaction_Result', 'customer_id']].drop_duplicates().reset_index(drop=True)
order_df
```

```python
import psycopg2
from sqlalchemy import create_engine

conn_string = 'postgresql://postgres:admin@127.0.0.1/Spi-projekt'

db = create_engine(conn_string)

try:
    order_table = order_df.to_sql('orders', db, if_exists='replace', index=False)
    customer_table = customer_df.to_sql('customers', db, if_exists='replace', index=False)
    product_table = product_df.to_sql('products', db, if_exists='replace', index=False)
    state_table = state_df.to_sql('states', db, if_exists='replace', index=False)
except Exception as ex:
    print(ex)
else:
    print("PostgreSQL tablice su kreirane i popunjene.")

```
