#### Izrada relacijskog modela i baze podataka

##### ER DIJAGRAM

![ER DIJAGRAM](ER_dijagram.png "ER DIJAGRAM")



```python
state_df = df[["customer_state"]].drop_duplicates().reset_index(drop=True)
state_df = state_df.rename(columns={"customer_state": "state_name"})
state_df.insert(0, "state_id", range(1, 1+len(state_df)))
state_df
```

```python
city_df = df[["customer_city", "customer_state"]].drop_duplicates().reset_index(drop=True)
city_df = city_df.rename(columns={"customer_city": "city_name", "customer_state": "state_name"})
city_df = pd.merge(city_df, state_df, on="state_name")
city_df = city_df[["city_name", "state_id"]].drop_duplicates().reset_index(drop=True)
city_df.insert(0, "city_id", range(1, 1+len(city_df)))
city_df
```

```python
df_all = pd.merge(city_df, state_df, on="state_id")
```

```python
customer_df = df[['customer_id', 'customer_zip_code_prefix',]].drop_duplicates().reset_index(drop=True)

# create city_df with unique city_id, city_name, and state_id values
city_df = df[['customer_zip_code_prefix', 'customer_city', 'customer_state']].drop_duplicates()
city_df['city_id'] = range(1, len(city_df)+1)

# merge customer_df and city_df dataframes on customer_zip_code_prefix and city_name columns
customer_df = pd.merge(customer_df, city_df, on='customer_zip_code_prefix')

# drop city_name column and reorder columns
customer_df = customer_df[['customer_id', 'customer_zip_code_prefix', 'city_id']]

customer_df
```

```python
# create seller_df dataframe with unique seller_id and seller_zip_code_prefix
seller_df = df[['seller_id', 'seller_zip_code_prefix']].drop_duplicates().reset_index(drop=True)

# create city_df dataframe with unique city_id, city_name, and state_id
city_df = df[['seller_zip_code_prefix', 'seller_city', 'seller_state']].drop_duplicates().reset_index(drop=True)
city_df['city_id'] = range(1, len(city_df)+1)

# merge seller_df and city_df dataframes on seller_zip_code_prefix column
seller_df = pd.merge(seller_df, city_df, on='seller_zip_code_prefix')

# keep only the desired columns and reorder them
seller_df = seller_df[['seller_id', 'seller_zip_code_prefix', 'city_id']]


seller_df
```

```python
review_df = df[['review_id', 'review_score', 'review_creation_date', 'review_answer_timestamp', 'customer_id','product_id',]].drop_duplicates().reset_index(drop=True)
review_df
```

```python
product_df = df[['product_id', 'product_category_name', 'product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'product_weight_g', 'product_length_cm','product_height_cm','product_width_cm']].drop_duplicates().reset_index(drop=True)
product_df
```

```python
payment_df = df[['payment_type', 'payment_installments', 'payment_value']].drop_duplicates().reset_index(drop=True)
payment_df['payment_id'] = payment_df.index
payment_df 
```

```python
order_df=pd.merge(df, payment_df, on=['payment_type', 'payment_installments', 'payment_value'])
order_df = order_df[['order_id', 'order_status', 'order_purchase_timestamp','order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date','price','freight_value','product_id','customer_id','payment_id','seller_id','review_id']].drop_duplicates().reset_index(drop=True)
order_df
```

```python
import psycopg2
from sqlalchemy import create_engine

conn_string = 'postgresql://postgres:admin@127.0.0.1/Spi-projekt'

db = create_engine(conn_string)

try:
    order_table = order_df.to_sql('orders', db, if_exists='replace', index=False)
    customer_table = customer_df.to_sql('customers', db, if_exists='replace', index=False)
    product_table = product_df.to_sql('products', db, if_exists='replace', index=False)
    review_table = review_df.to_sql('reviews', db, if_exists='replace', index=False)
    payment_table = payment_df.to_sql('payments', db, if_exists='replace', index=False)
    seller_table = seller_df.to_sql('sellers', db, if_exists='replace', index=False)
    city_table = city_df.to_sql('cities', db, if_exists='replace', index=False)
    state_table = state_df.to_sql('states', db, if_exists='replace', index=False)
except Exception as ex:
    print(ex)
else:
    print("PostgreSQL tablice su kreirane i popunjene.")

```

## Checkpoint 3 

#### Izrada dimenzijskog modela i 

##### DIMENZIJSKI MODEL

![DIMENZIJSKI_MODEL](DIM_model.png "DIMENZIJSKI MODEL")


```python
import numpy as np
```

##### Dimenzijska tablica Customer


```python
dim_customer_df = pd.merge(state_df, city_df, on="state_id"
                     ).merge(customer_df, on="city_id")
dim_customer_df['dim_customer_id'] = dim_customer_df.index + 1
dim_customer_df['start_date'] = pd.Timestamp.today()
dim_customer_df['end_date'] = np.nan
dim_customer_df = dim_customer_df[['dim_customer_id', 'customer_id', 'customer_zip_code_prefix','city_name', 'state_name', 'start_date', 'end_date']]

dim_customer_df
```

##### Dimenzijska tablica Seller


```python
dim_seller_df = pd.merge(state_df, city_df, on="state_id"
                     ).merge(seller_df, on="city_id")
dim_seller_df['dim_seller_id'] = dim_seller_df.index + 1
dim_seller_df['start_date'] = pd.Timestamp.today()
dim_seller_df['end_date'] = np.nan
dim_seller_df = dim_seller_df[['dim_seller_id', 'seller_id', 'seller_zip_code_prefix','city_name', 'state_name', 'start_date', 'end_date']]

dim_seller_df
```

##### Dimenzijska tablica Review


```python
dim_review_df = review_df.copy()

dim_review_df['dim_review_id'] = dim_review_df.index + 1
dim_review_df['start_date'] = pd.Timestamp.today()
dim_review_df['end_date'] = np.nan
dim_review_df = dim_review_df[['dim_review_id', 'review_id', 'review_score','review_creation_date', 'review_answer_timestamp', 'start_date', 'end_date']]

dim_review_df
```

##### Dimenzijska tablica Product


```python
dim_product_df = product_df.copy()

dim_product_df['dim_product_id'] = dim_product_df.index + 1
dim_product_df['start_date'] = pd.Timestamp.today()
dim_product_df['end_date'] = np.nan
dim_product_df = dim_product_df[['dim_product_id', 'product_id', 'product_category_name','product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm', 'start_date', 'end_date']]

dim_product_df
```

##### Dimenzijska tablica Payment


```python
dim_payment_df = payment_df.copy()

dim_payment_df['dim_payment_id'] = dim_payment_df.index + 1
dim_payment_df['start_date'] = pd.Timestamp.today()
dim_payment_df['end_date'] = np.nan
dim_payment_df = dim_payment_df[['dim_payment_id', 'payment_id', 'payment_type','payment_installments', 'payment_value', 'start_date', 'end_date']]

dim_payment_df
```

##### Dimenzijska tablica Order_date


```python
dim_order_date_df = order_df.copy()

dim_order_date_df = dim_order_date_df[['order_purchase_timestamp','order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']].drop_duplicates()
dim_order_date_df['dim_order_date_id'] = dim_order_date_df.index + 1

dim_order_date_df
```

##### Dimenzijska tablica Order


```python
ft_order_df = order_df.merge(dim_customer_df, on='customer_id'
                            ).merge(dim_seller_df, on='seller_id'
                            ).merge(dim_review_df, on='review_id'
                            ).merge(dim_product_df, on='product_id'
                            ).merge(dim_payment_df, on='payment_id'
                            ).merge(dim_order_date_df, on=['order_purchase_timestamp','order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date'])
ft_order_df['ft_order_id'] = ft_order_df.index + 1
ft_order_df = ft_order_df[['ft_order_id', 'order_id', 'order_status', 'freight_value', 'dim_customer_id', 'dim_seller_id', 'dim_review_id', 'dim_product_id', 'dim_payment_id', 'dim_order_date_id']].drop_duplicates()
ft_order_df
```

##### Spajanje sa skladištem i popunjavanje podacima

```python
import psycopg2
from sqlalchemy import create_engine

conn_string = 'postgresql://postgres:admin@127.0.0.1/SPI-dim-projekt'

db = create_engine(conn_string)

try:
    ft_order_table = ft_order_df.to_sql('orders', db, if_exists='replace', index=False)
    dim_customer_table = dim_customer_df.to_sql('customers', db, if_exists='replace', index=False)
    dim_product_table = dim_product_df.to_sql('products', db, if_exists='replace', index=False)
    dim_review_table = dim_review_df.to_sql('reviews', db, if_exists='replace', index=False)
    dim_payment_table = dim_payment_df.to_sql('payments', db, if_exists='replace', index=False)
    dim_seller_table = dim_seller_df.to_sql('sellers', db, if_exists='replace', index=False)
    dim_order_date_table = dim_order_date_df.to_sql('order_date', db, if_exists='replace', index=False)

except Exception as ex:
    print(ex)
else:
    print("PostgreSQL tablice su kreirane i popunjene.")

```
